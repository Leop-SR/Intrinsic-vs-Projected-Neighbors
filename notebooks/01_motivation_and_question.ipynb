## Introduction

High-dimensional datasets are common in modern data analysis, but analyzing them directly can be challenging due to issues related to distance geometry. As dimensionality increases, distances between points tend to become less informative, which complicates tasks based on similarity or proximity.

Dimensionality reduction techniques, such as Principal Component Analysis (PCA), are often used to address these challenges by performing linear projections of the data into lower-dimensional spaces.

While PCA is known to preserve global variance through the eigenvalues and eigenvectors of the data covariance matrix, it is less clear how this linear projection affects local geometric structures and how much information is lost in the process, particularly in terms of nearest-neighbor relationships.

Understanding how neighborhood structure changes under PCA is therefore important for distance-based algorithms, especially k-nearest neighbors (KNN).


## Research Questions

- How does the nearest-neighbor structure of the data change after applying PCA?
- To what extent are k-nearest neighbors preserved when projecting data into lower-dimensional subspaces?
- What implications does this have for distance-based algorithms such as KNN?


## Hypotheses

- Projecting high-dimensional data into a lower-dimensional subspace using PCA may make distance relationships more informative by reducing the effects of high-dimensional geometry.
- Despite preserving global variance, PCA discards information contained in the null space of the projection, which may alter local neighborhood relationships.
- As a consequence, nearest-neighbor structures are expected to change after projection, with the degree of distortion increasing as dimensionality is further reduced.
